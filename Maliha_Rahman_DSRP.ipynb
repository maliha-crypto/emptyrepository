```python
import pandas as pd
```

```python
# Repository + upload the data set
df = pd.read_csv("health_and_nutrition_combined_clean.csv")
df.head()
```

```python
# Change the Column names
# Change the codes to what they actually represent
# Drop uneccesary columns
# Check misisng values/data
```

```python
# Rename the columns
df = df.rename(columns = {
    "SEQN": "Participant ID",
    "RIAGENDR": "Gender",
    "RIDRETH3": "Race Ethnicity",
    "RIDAGEYR": "Age",
    "BMXHT": "Height",
    "BMXBMI": "BMI",
    "BMXWT": "Weight",
    "RIDEXPRG": "Pregnancy Status",
    "WTMEC4YR": "Survey Weight"
})
df.columns
```

```python
# Drop the "Unnamed: 0" column
df = df.drop(columns=["Unnamed: 0"])
df.head()
```

```python
# Change the codes to what they actually mean, gender
df["Gender"] = df["Gender"].replace({1: "Male", 2: "Female"})
df.head()
```

```python
# Now same thing for "Race Ethnicity"
# Check which ones are actually used in the dataset
df["Race Ethnicity"].unique()
```

```python
df["Race Ethnicity"] = df["Race Ethnicity"].replace({
    1: "Mexican American",
    3: "Non-Hispanic White",
    4: "Non-Hispanic Black",
    6: "Non-Hispanic Asian",
    7: "Other Race, Including Multiracial"
})
df.head()
```

```python
# Check NaNs
df.isnull().sum()
```

```python
# Drop "Survey Weight"
df = df.drop(columns=["Survey Weight"])
df.head()
```

```python
# Drop "Pregnancy Status"
df = df.drop(columns=["Pregnancy Status"])
df.head()
```

```python
df.columns
```

```python
# Check for duplicates
df.duplicated().sum()
```

```python
# Check for missing values
df.isnull().sum()
```

```python
# Dropping the missing values in Height, BMI, and Weight
df = df.dropna()
df.isnull().sum()
```
